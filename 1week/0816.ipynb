```python
import numpy as np
x = np.array([2,4,6,8])
y = np.array([81,93,91,97])
mx = np.mean(x)
my = np.mean(y)
print("x의 평균값: ", mx)
print("y의 평균값: ", my)
divisor = sum([(i-mx)**2 for i in x])
```

```python
def top(x, mx, y, my):
  d = 0
  for i in range(len(x)):
    d += (x[i] - mx) * (y[i] - my)
    return d
dividend = top(x, mx, y ,my)

print("분모: ",divisor)
print("분자: ",dividend)

a = dividend / divisor
b = my - (mx*a)

print("기울기 a =",a)
print("y절편 b =",b)
```

```python
fake_a = 3
fake_b  = 76

def predict(x):
  return fake_a * x + fake_b

predict_result = []

for i in range(len(x)):
  predict_result.append(predict(x[i]))
  print("공부시간 = %.f, 실제점수=%.f, 예측점수=%.f"%(x[i],y[i],predict(x[i])))

n = len(x)
def mse(y,y_pred):
  return (1/n) * sum((y-y_pred)**2)

print("평균 제곱 오차:" + str(mse(y, predict_result)))
```

# 파이썬 코딩으로 확인하는 다중 선형 **회귀**

```python
import numpy as np
import matplotlib.pyplot as plt


x= np.array([2, 4, 6, 8])
y= np.array([81, 93, 91, 97])
#데이터의 분포를 그래프로
#plt.scatter(x, y)
#plt.show()
#기울기 a의 값과 절편 b의 값을 초기화 (중요)
a=0
b=0
#학습률
lr= 0.03
#몇 번 반복할지
epochs = 2001
n=len(x)
#경사 하강법 시작
for i in range(epochs):
  y_pred= a* x + b
  error= y - y_pred
  a_diff = (2/n) * sum(-x * (error))
  b_diff = (2/n) * sum(-(error))
  a = a-lr * a_diff
  b = b-lr * b_diff
  if i%100 == 0: #100번 반복될 때마다 현재의 a값, b값 출력
    print(f"epoch ={i}, 기울기 = {a:4f}, 절편={b:4f}")
#최종 a값을 기울기, b값을 y 절편에 대입해 그래프화
y_pred = a * x + b
plt.scatter(x, y)
plt.plot(x, y_pred, 'r')
plt.show()
```

# 선형회귀,다중 선형회귀 *모델*

```python
import numpy as np
import matplotlib.pyplot as plt

from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense

x = np.array([2,4,6,8])
y = np.array([81,93,91,97])

model = Sequential()

model.add(Dense(1, input_dim=1,activation='linear'))

model.compile(optimizer='sgd',loss='mse')

model.fit(x, y, epochs=100)

plt.scatter(x,y)
plt.plot(x, model.predict(x),'r')
plt.show()

hour =  np.array([7])
prediction = model.predict([hour])
print("%.f시간을 공부할 경우의 예상 점수는 %.02f점입니다." % (hour, prediction))
```

```python
x = np.array([[2,0],[4,4],[6,2],[8,3]])
y = np.array([81,93,91,97])

model = Sequential()

model.add(Dense(1, input_dim=2,activation='linear'))

model.compile(optimizer='sgd',loss='mse')

model.fit(x, y, epochs=100)

hour = 7
private_class = 4
prediction = model.predict(np.array([[hour,private_class]]))
print("%.f시간을 공부하고 %.f시간의 과외를 받을 경우, 예상 점수는 %.02f점입니다." % (hour,private_class,prediction[0][0]))
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error, r2_score
```

```python
#X, Y 분할하기
california_df['PRICE'] = california.target
Y = california_df['PRICE']
X = california_df.drop(['PRICE'], axis = 1, inplace = False)
print('보스톤 주택 가격 데이터셋 크기: ', california_df.shape)
california_df.info()
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size = 0.3, random_state = 156)
#선형 회귀 분석 : 모델 생성
lr = LinearRegression()
#선형 회귀 분석 : 모델 훈련
lr.fit(X_train, Y_train)
#선형 회귀 분석 : 평가 데이터에 대한 예측 수행 -> 예측 결과 Y_predict 구하기
Y_predict = lr.predict(X_test)
```

```python
mse = mean_squared_error(Y_test, Y_predict)
rmse = np.sqrt(mse)
print('MSE : {0:.3f}, RMSE : {1:.3f}'.format(mse, rmse))
print('R^2(Variance score) : {0:.3f}'.format(r2_score(Y_test, Y_predict)))
```

```python
print('Y 절편 값: ', lr.intercept_)
print('회귀 계수 값: ', np.round(lr.coef_, 1))
```
