# **타이타닉호 분석 `seaborn`**

```python
import seaborn as sns
import pandas as pd

titanic = sns.load_dataset('titanic')
titanic
```

```python
titanic.isnull().sum()  #NaN(NULL)값의 갯수
```

```python
# .fillna(titanic['age'].median()은 age 열에서 결측값을 age 열의 중앙값(median)으로 채웁니다.
titanic['age'] = titanic['age'].fillna(titanic['age'].median())
#titanic 데이터셋의 embarked 열에서 각 값이 얼마나 자주 등장하는지(최빈값)를 계산  최빈값: '최'고로 '빈'도가 많은 '값'
titanic['embarked'].value_counts()
```

```python
titanic['embark_town'] = titanic['embark_town'].fillna('Southampton')
titanic['deck'].value_counts()
```

```python
titanic['deck'] = titanic['deck'].fillna('C')
titanic.isnull().sum()
```

```python
titanic.survived.value_counts()
```

```python
titanic.who.value_counts()
```

```python
titanic.info()
```

```python
import matplotlib.pyplot as plt
f,ax= plt.subplots(1, 2, figsize= (10, 5))
titanic['survived'][titanic['sex'] == 'male'].value_counts().plot.pie(explode = [0,0.1], autopct= '%1.1f%%', ax = ax[0], shadow = True)
titanic['survived'][titanic['sex'] == 'female'].value_counts().plot.pie(explode = [0,0.1], autopct= '%1.1f%%', ax = ax[1], shadow = True)
ax[0].set_title('Survived (Male)')
ax[1].set_title('Survived (Female)')
plt.show()
```

```python
sns.countplot(x='pclass', hue = 'survived', data = titanic)

plt.title('Pclass vs Survived')
plt.show()
```

```python
titanic['sex'] = titanic['sex'].map({'male':0,'female':1})
titanic['who'] = titanic['who'].map({'man':1,'woman':0, 'child':-1})
titanic_corr = titanic.corr(method='pearson' ,numeric_only=True)
titanic_corr
```

```python
sns.pairplot(titanic, hue = 'survived')
plt.show()
```

```python
sns.catplot(x = 'pclass', y = 'survived', hue = 'sex', data = titanic, kind = 'point')
plt.show()
```

```python
def category_age(x):
  if x < 10:
    return 0
  elif x < 20:
    return 1
  elif x < 30:
    return 2
  elif x < 40:
    return 3
  elif x < 50:
    return 4
  elif x < 60:
    return 5
  elif x < 70:
    return 6
  else:
    return 7
```

```python
titanic['age2'] = titanic['age'].apply(category_age)
titanic['sex'] = titanic['sex'].map({'male':1, 'female':0})
titanic['family'] = titanic['sibsp'] + titanic['parch'] + 1
titanic.to_csv('/content/drive/MyDrive/Dosan/train.csv', index = False)
heatmap_data = titanic[['survived', 'sex', 'age2', 'family', 'pclass', 'fare']]
colormap = plt.cm.RdBu
sns.heatmap(heatmap_data.astype(float).corr(), linewidths = 0.1, vmax = 1.0, square = True, cmap = colormap, linecolor = 'white', annot = True,
annot_kws = {"size": 10})
plt.show()
```

# **모두의 딥러닝 12장**

```python
import pandas as pd

!git clone http://github.com/taehojo/data.git

df = pd.read_csv('./data/iris3.csv')

df.head()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

sns.pairplot(df, hue='species');
plt.show()
```

```python
X = df.iloc[:,0:4]
y = df.iloc[:,4]

print(X[0:5])
print(y[0:5])
```

```python
y = pd.get_dummies(y)

print(y[0:5])
y
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
import numpy as np
!git clone https://github.com/taegojo/data.git

model = Sequential()
model.add(Dense(12, input_dim=4, activation='relu'))
model.add(Dense(8,activation='relu'))
model.add(Dense(3,activation='softmax'))
model.summary()

model.compile(loss='categorical_crossentropy', optimizer='adam', metrics=['accuracy'])
```

```python
history = model.fit(X, y, epochs = 50, batch_size = 5)
```
