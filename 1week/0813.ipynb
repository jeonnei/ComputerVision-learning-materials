# 단순한 시퀀스는
`range()`함수의 실수버전인 `arrange(from,to,step)` 이나

`linspace(from,to,npoints)`를 사용하면 편리

또한 단위행렬을 위한 `eye(n)`함수를 제공한다.

```python
print(np.arange(1,2,0.1))

print(np.arange(10))

print(np.arange(10.))

print(np.linspace(0.,20.,11))

print(np.eye(3))
```
