# 06차시 과제

### Q1. "Data"와 "Information"의 차이를 정의하고, 그 차이를 설명하는 예시를 하나 들어 설명하시오.

Data는 단순한 사실이나 측정값을 의미하며, Information는 이러한 데이터를 분석하고 해석하여 맥락이나 의미를 부여한 상태를 말합니다. 예를 들어, 날씨 데이터가 수집된 후, 그 데이터를 분석하여 특정 날의 기온, 습도, 강수 확률 등으로 해석하면 정보가 됩니다.

### Q2. 다음 중 Data에 대한 설명으로 옳은 것을 고르시오.

1. 데이터를 특정 맥락에서 처리하여 얻어진 의미 있는 내용이다.
2. 가공되지 않은 원시적인 사실이나 숫자, 관찰된 값이다.
3. 특정 상황에서 의사 결정을 돕기 위해 사용되는 해석된 자료이다.
4. 정보를 해석한 후 얻어진 데이터이다.

### Q7. Cost function이 무엇이고 언제 무엇을 위해 사용하는지에 대해 서술하시오.


비용 함수(Cost Function)는 머신러닝 모델의 예측값과 실제값 간의 차이를 수치적으로 나타내는 함수입니다. 이는 모델의 학습 성과와 예측 정확도를 평가하는 데 사용됩니다.

### Q8. 지도학습과 비지도학습의 차이점에 대해 서술하시오.

정답을 제공받으면서 학습하는 방법은 지도학습이며, 정답 없이 학습하는 방법은 비지도학습입니다.
